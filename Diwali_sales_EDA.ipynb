*To see the count of male and females present in our data*

```python
ax=sns.countplot(x='Gender',data=df,hue='Gender')
for bars in ax.containers:
  ax.bar_label(bars)
```

*Arrange the data as per gender distribution and see how much amount spend by males and females*

```python
sales_gen=df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sales_gen
```

```python
sales_gen=df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
a=sns.barplot(x='Gender',y='Amount',data=sales_gen,hue ='Gender')
for bars in a.containers:
  a.bar_label(bars)
```

*We can see that the amount spend by females are greater than males*

```python
ax=sns.countplot(data=df,x='Age Group', hue='Gender')
for bars in ax.containers:
  ax.bar_label(bars)
```

*The age_group between 26-35 had done most sales*

```python
#Total amount VS Age Group
```

```python
sales_age=df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sales_age
```

```python
sns.barplot(x='Age Group',y='Amount',data=sales_age,hue='Age Group')
```

*The age_group 26-35 spend most of the amount in the sales *

```python
# Total number of Orders from the top 10 states
sales_states=df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
sns.set(rc={'figure.figsize':(15,5)})
b=sns.barplot(x='State',y='Orders',data=sales_states,hue='State')
for bars in b.containers:
  b.bar_label(bars)
```

*The top orders are given from the states: UP, Maharashtra, Karnataka etc.*

```python
 #Total number of sales from Top 10 Sales
 state_spend=df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
c=sns.barplot(x='State',y='Amount',data=state_spend,hue='State')
```

```python
sns.set(rc={'figure.figsize': (5,5)})
d=sns.countplot(x='Marital_Status',data=df,hue='Marital_Status')

for bars in d.containers:
  d.bar_label(bars)

```

```python
marital_amount=df.groupby(['Marital_Status','Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Marital_Status',y='Amount',hue='Gender',data=marital_amount)
```

```python
sns.set(rc={'figure.figsize':(15,5)})
occ_plot=sns.countplot(data=df,x='Occupation',hue='Occupation')
for bars in occ_plot.containers:
  occ_plot.bar_label(bars)
```

```python
occu_sales=df.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Occupation',y='Amount',data=occu_sales,hue='Occupation')
```

*The top 3 Occupations which spend most of he amount are: IT sector, Healthcare, Aviation *

```python
sns.set(rc={'figure.figsize':(25,5)})
product_plot=sns.countplot(x='Product_Category',data=df,hue='Product_Category')
for bars in product_plot.containers:
  product_plot.bar_label(bars)
```

```python
product_sales=df.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.barplot(x='Product_Category',y='Amount',data=product_sales,hue='Product_Category')
```

*The top 3 amount spend by the Categories are: Food, Clothing and apparel and Electronics*

**Conclusions**
*Married women of age-group 26-35 from UP, Maharashtra and Karnataka working in IT, Healthcare and aviation are more likely to buy products from food, clothing and apparel and electronics.*
